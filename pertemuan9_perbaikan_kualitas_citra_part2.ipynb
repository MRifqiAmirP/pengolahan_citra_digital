```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv.imread("Gambar/Pertemuan 9/noise_image.png")
plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
plt.axis('off')
```

```python
# PENAPIS RERATA

kernel = np.array([[1, 1, 1],
                   [1, 1, 1],
                   [1, 1, 1]])/9

print(kernel)
```

```python
kernels = np.ones((3, 3), np.float32)/9
print(kernels)
```

```python
blur_image = cv.filter2D(image, -1, kernels)
plt.imshow(cv.cvtColor(blur_image, cv.COLOR_BGR2RGB))
plt.axis('off')
```

```python
kernel_hpf = [None] * 6

# KERNEL 1
kernel_hpf[0] = np.array([[-1, -1, -1],
                       [-1, 8, -1],
                       [-1, -1, -1]])

# KERNEL 2
kernel_hpf[1] = np.array([[-1, -1, -1],
                       [-1, 9, -1],
                       [-1, -1, -1]])

# KERNEL 3
kernel_hpf[2] = np.array([[0, -1, 0],
                       [-1, 5, -1],
                       [0, -1, 0]])

# KERNEL 4
kernel_hpf[3] = np.array([[1, -2, 1],
                       [-2, 5, -2],
                       [1, -2, 1]])

# KERNEL 5
kernel_hpf[4] = np.array([[1, -2, 1],
                       [-2, 4, -2],
                       [1, -2, 1]])

# KERNEL 6
kernel_hpf[5] = np.array([[0, 1, 0],
                       [1, -4, 1],
                       [0, 1, 0]])

for i in range(len(kernel_hpf)):
    print(kernel_hpf[i])
```

```python
for i in range(len(kernel_hpf)):
    sharp_img = cv.filter2D(image, -1, kernel_hpf[i])
    plt.subplot(2, 3, i + 1)
    plt.imshow(cv.cvtColor(sharp_img, cv.COLOR_BGR2RGB))
    plt.axis('off')
    plt.title(f'Kernel {i+1}')

plt.tight_layout()
plt.show()
```

```python
averaging = cv.blur(image, (21, 21))
gaussian = cv.GaussianBlur(image, (21,21), 0)
median = cv.medianBlur(image, 5)
bilateral = cv.bilateralFilter(image, 9, 350, 350)

image_repair = [averaging, gaussian, median, bilateral]
metode_names = ['Averaging', 'Gaussian Blur', 'Median Blur', "Bilateral"]
```

```python
for i in range(len(image_repair)):
    plt.subplot(2, 2, i + 1)
    plt.imshow(cv.cvtColor(image_repair[i], cv.COLOR_BGR2RGB))
    plt.title(f'{metode_names[i]}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
