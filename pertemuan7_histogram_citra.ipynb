```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv.imread("Gambar/gambar-kucing.jpg", 0)
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
histogram = cv.calcHist([image], [0], None, [256], [0, 256])
```

```python
plt.figure()
plt.title("Histogram")
plt.xlabel("Intensitas piksel")
plt.ylabel("Jumlah piksel")
plt.plot(histogram, color='black')
plt.xlim([0, 256])
plt.show()
```

```python
# GAMBAR BERWARNA

image_clr = cv.imread("Gambar/gambar-kucing.jpg")
plt.imshow(cv.cvtColor(image_clr, cv.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
blue_channel = image_clr[:, :, 0]

plt.imshow(blue_channel, cmap="gray")
plt.axis('off')
plt.title("Blue Channel")
plt.show()
```

```python
green_channel = image_clr[:, :, 1]

plt.imshow(green_channel, cmap="gray")
plt.axis('off')
plt.title("Green Channel")
plt.show()
```

```python
red_channel = image_clr[:, :, 2]

plt.imshow(red_channel, cmap="gray")
plt.axis('off')
plt.title("Red Channel")
plt.show()
```

```python
histogram_blue = cv.calcHist([image_clr], [0], None, [256], [0, 256])
histogram_green = cv.calcHist([image_clr], [1], None, [256], [0, 256])
histogram_red = cv.calcHist([image_clr], [2], None, [256], [0, 256])
```

```python
# Menampilkan histogram
plt.figure(figsize=(10, 5))
plt.title("Histogram of Color Channels")
plt.xlabel("Pixel Intensity")
plt.ylabel("Jumlah Piksel")
plt.xlim([0, 256])

# Plot histogram untuk masing-masing channel
plt.plot(histogram_blue, color='blue', label='Blue Channel')
plt.plot(histogram_green, color='green', label='Green Channel')
plt.plot(histogram_red, color='red', label='Red Channel')

# Menambahkan legenda
plt.legend()
plt.grid()  # Menambahkan grid untuk memudahkan pembacaan
plt.show()
```

```python
plt.subplot(3, 1, 1)
plt.plot(histogram_blue, color='blue')
plt.title('Histogram of Blue Channel')
plt.xlim([0, 256])
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.grid()

plt.subplot(3, 1, 2)
plt.plot(histogram_green, color='green')
plt.title('Histogram of Green Channel')
plt.xlim([0, 256])
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.grid()

plt.subplot(3, 1, 3)
plt.plot(histogram_red, color='red')
plt.title('Histogram of Red Channel')
plt.xlim([0, 256])
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.grid()

plt.tight_layout()
plt.show()
```

```python
dark_image = cv.imread("Gambar/Pertemuan 7/Hutan-Gelap.jpeg", 0)
plt.imshow(dark_image, cmap="gray")
plt.axis('off')
plt.show()
```

```python
histogram_before = cv.calcHist([dark_image], [0], None, [256], [0, 256])

plt.figure()
plt.xlabel("Jumlah Piksel")
plt.ylabel("Intensitas")
plt.plot(histogram_before)
plt.xlim([0, 256])
plt.show()
```

```python
equalized_image = cv.equalizeHist(dark_image)

plt.imshow(equalized_image, cmap="gray")
```

```python
histogram_after = cv.calcHist([equalized_image], [0], None, [256], [0, 256])

plt.figure()
plt.xlabel("Jumlah piksel")
plt.ylabel("Intensitas")
plt.plot(histogram_after)
plt.xlim([0, 256])
plt.show()
```
